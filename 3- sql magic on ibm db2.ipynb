```python
%%sql
select e.f_name, e.l_name, jh.start_date ,jh.dept_id 
from employees as e, job_history as jh
where e.emp_id = jh.empl_id
and jh.dept_id = 5
;
```

```python
%%sql
select e.f_name, e.l_name, e.job_id, jh.start_date , jh.dept_id
from employees as e inner join job_history as jh
on e.emp_id = jh.empl_id
where jh.dept_id = 5
;
```

```python
%%sql
select a.f_name, a.l_name, a.job_id, j.job_title, a.start_date , a.dept_id
from (select e.f_name, e.l_name, e.job_id, jh.start_date , jh.dept_id
    from employees as e inner join job_history as jh
    on e.emp_id = jh.empl_id
    where jh.dept_id = 5) as a inner join jobs as j
    on a.job_id=j.job_ident
;
```

```python
%%sql
select e.emp_id, e.l_name, d.dept_id_dep, d.dep_name, e.b_date
from employees as e left join departments as d
on e.dep_id = d.dept_id_dep
;
```

```python
%%sql
select e.emp_id, e.l_name, e.dep_id, b.dep_name, e.b_date
from employees as e left join ( select e.emp_id, e.l_name, d.dept_id_dep, d.dep_name, e.b_date
    from employees as e left join departments as d
    on e.dep_id = d.dept_id_dep
    where Year(e.b_date) < '1980') as b 
on e.emp_id = b.emp_id
;
```

```python
%%sql
select e.f_name, e.l_name, d.dep_name
from employees as e full outer join departments as d
on e.dep_id=d.dept_id_dep
;
```

```python
%%sql
select e.f_name, e.l_name, d.dep_name , d.dept_id_dep, e.sex
from employees as e left outer join departments as d
on e.dep_id=d.dept_id_dep and e.sex='F'
;
```
